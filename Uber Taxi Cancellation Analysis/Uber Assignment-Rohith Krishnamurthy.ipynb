```python
# Import the numpy and pandas packages

import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the csv file using 'read_csv'

uber = pd.read_csv('Uber Request Data.csv')
uber
```

```python
#Data cleaning and changing the Request timestamp and Drop timestamp to proper format
uber['Request timestamp'] =  pd.to_datetime(uber['Request timestamp'],dayfirst=True)
uber['Drop timestamp'] =  pd.to_datetime(uber['Drop timestamp'],dayfirst=True)
uber
```

```python
#Create a seperate column for the request and drop hour from the corresponding timestamp and extract the hour from the timestamp
uber['req_hour'] = uber['Request timestamp'].map(lambda x:pd.to_datetime(x).hour)
uber['drop_date'] = uber['Drop timestamp'].map(lambda x:pd.to_datetime(x).day)
uber['req_date'] = uber['Request timestamp'].map(lambda x:pd.to_datetime(x).day)
uber
```

```python
#Create 6 different timeslots for the 24 hours in a day and make a new column Time_of_Day from hours column
uber['Time_of_Day'] = pd.cut(uber.req_hour, bins=[0,4,8,12,16,20,24], labels=["late_night","early_morning","morning", "afternoon", "evening","night"])
uber
```

```python
#Create a plot to show the status of all the cab requests
ax = sns.countplot(x="Status",data=uber)
```

```python
#Create a plot to show the count for the two pickup points
ax = sns.countplot(x="Pickup point",data=uber)
```

```python
#Create a plot to show the count of requests during each time slot from both the pickup points together
ax = sns.countplot(x="Time_of_Day",data=uber)
```

```python
#Create a pivot table to get the count of the various status of the trips on each of the 5 days in the dataset
uber1 = uber.pivot_table(index=["req_date"],columns='Status', values=["Request id"],aggfunc=len)
uber1
```

```python
#Create a plot to get the count of the various status of the trips on each of the 5 days in the dataset
uber1.plot.bar(figsize=(8,8),stacked = True)
```

```python
#Create a pivot table to get the count of total requests from both pickup points on each of the 5 days in the dataset
uber2 = uber.pivot_table(index=["req_date"],columns='Pickup point', values=["Request id"],aggfunc=len)
uber2
```

```python
#Create a plot to get the count of total requests from both pickup points on each of the 5 days in the dataset

uber2.plot.bar(figsize=(12,8),stacked = True)
bar_width = 2.5
```

```python
#Create a pivot table to get the count of total requests and their status in all the 6 time slots of the day

uber3 = uber.pivot_table(index=["Status"],columns='Time_of_Day', values=["Request id"],aggfunc=len)
uber3
```
